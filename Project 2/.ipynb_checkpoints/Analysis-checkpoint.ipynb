## Problem Statement
The rise of spam comments on YouTube has become a significant challenge, affecting user experience and platform integrity. The objective of this project is to develop a predictive model that can accurately classify YouTube comments as either spam or legitimate. Using the provided [Kaggle Dataset](https://www.kaggle.com/datasets/ahsenwaheed/youtube-comments-spam-dataset/data), the task is to create models that maximize predictive performance to effectively distinguish between spam and non-spam comments. This problem aims to leverage machine learning techniques to address a critical issue in content moderation.
## Data
The dataset is linked above. It consists of 1,956 real messages extracted from five videos that were among the 10 most viewed on the collection period.

Model: Since I want to make this as accurate as possible and not finding the best features, I was thinking of using a [TfidfVectorizer](https://scikit-learn.org/stable/modules/generated/sklearn.feature_extraction.text.TfidfVectorizer.html) to convert the content into an array of word importance based on frequency. I will go more into it's formula when I do research on it. The resulting array will then be fed through a SVM or Logistic Regression model to be evaulated. 

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report

X = data['CONTENT']
y = data['CLASS']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42, test_size = 0.2)

tfidf_vectorizer = TfidfVectorizer(stop_words='english', use_idf = True)

X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)

X_test_tfidf = tfidf_vectorizer.transform(X_test)
```

```python
from tqdm import tqdm
df_list = []
for vector in tqdm(range(X_train_tfidf.shape[0])):
    val = X_train_tfidf[vector]
    df = pd.DataFrame(val.T.todense(), 
                  index=tfidf_vectorizer.get_feature_names_out(),
                  columns=["tfidf"]).sort_values(by='tfidf', ascending = False)
    df_list.append(df)
    
```

```python
tdif_df = pd.concat(df_list)

tdif_df.groupby(tdif_df.index).sum().sort_values(by='tfidf', ascending=False).head(10)
```

```python
log = LogisticRegression()

log.fit(X_train_tfidf, y_train)
```

```python
pd.DataFrame({'word': tfidf_vectorizer.get_feature_names_out(), 'coef': log.coef_.flatten()}).sort_values(by='coef',ascending=False)
    
    
```

```python
y_pred = log.predict(X_test_tfidf)
matrix = confusion_matrix(y_test, y_pred)
```

```python
plt.style.use('default')
ConfusionMatrixDisplay.from_estimator(log, X_test_tfidf, y_test)
plt.title('Logistic Regression')
plt.show()
```

```python
print(classification_report(y_test, y_pred))
```

```python
svc = SVC()

# Attempt to tune the kernel
param_grid = {'C': [0.1, 1, 10, 100, 1000],  
              'degree': [1, 2, 3, 4], 
              'kernel': ['rbf', 'poly']}  

grid = GridSearchCV(svc, param_grid, refit = True, verbose = 1)
```

```python
grid.fit(X_train_tfidf, y_train)
```

```python
print(grid.best_params_) 
print(grid.best_score_) 
```

```python
y_pred = grid.predict(X_test_tfidf)
```

```python
fig, ax = plt.subplots(1, 2, figsize = (10, 6))
ConfusionMatrixDisplay.from_estimator(log, X_test_tfidf, y_test, ax = ax[0], colorbar = False)
ConfusionMatrixDisplay.from_estimator(grid, X_test_tfidf, y_test, ax = ax[1], colorbar = False)

ax[0].set_title('Support Vector Classifier')
ax[1].set_title('Logistic Regression')

plt.show()
```

```python
print(classification_report(y_test, y_pred))
```
