## Problem Statement
The rise of spam comments on YouTube has become a significant challenge, affecting user experience and platform integrity. The objective of this project is to develop a predictive model that can accurately classify YouTube comments as either spam or legitimate. Using the provided [Kaggle Dataset](https://www.kaggle.com/datasets/ahsenwaheed/youtube-comments-spam-dataset/data), the task is to create models that maximize predictive performance to effectively distinguish between spam and non-spam comments. This problem aims to leverage machine learning techniques to address a critical issue in content moderation.
## Data
The dataset is linked above. It consists of 1,956 real messages extracted from five videos that were among the 10 most viewed on the collection period.

```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('Youtube-Spam-Dataset.csv')

data
```

```python
data.columns
```

```python
data.isna().sum()
```

```python
data.shape
```

```python
data['CLASS'].value_counts() / len(data)
```

```python
spam = data.query('CLASS == 1')['CONTENT'].str.cat(sep=' ')
non_spam = data.query('CLASS == 0')['CONTENT'].str.cat(sep=' ')
```

```python
wc = WordCloud().generate(text= non_spam)
plt.imshow(wc)
plt.title('Non Spam')
```

```python
wc = WordCloud().generate(text= spam)
plt.imshow(wc)
plt.title('SPAM')
```

```python
data['DATE'] = pd.to_datetime(data['DATE'], format='%Y-%m-%dT%H:%M:%S.%f', errors='coerce').fillna(pd.to_datetime(data['DATE'], errors='coerce'))


data['hour'] = data['DATE'].dt.hour
data['minute'] = data['DATE'].dt.minute
data['second'] = data['DATE'].dt.second
```

```python
data.isna().sum()
```

```python
data.groupby('CLASS')[['hour', 'minute', 'second']].mean()
```

```python
data['content_length'] = data['CONTENT'].str.split(' ').str.len()
```

```python
sns.regplot(x='hour', y='content_length', data=data, scatter=True, fit_reg=True, line_kws=dict(color='red'))

plt.title('Does Content Length Get Longer Throughout the Day?')

plt.text(26, 120, 'No correlation.\n\nMay not be a useful feature when\npredicting spam.')
plt.show()
```

```python
sns.kdeplot(x='content_length', hue = 'CLASS', data = data, fill = True)
plt.title('Density of Content Lengths for Non-Spam vs. Spam')
plt.show()
```
